### KNN: K-최근접 이웃 알고리즘
- KNN(k-Nearest Neighbors algorithm)
- 도미/빙어 데이터를 합치기

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
len(length) # 35 + 14
```

```python
len(weight) # 35 + 14
```

```python
for l, w in zip(length, weight):
    print(f'{l, w}')
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]    # zip은 length, weight를 묶어줌
print(fish_data)
```

각 데이터에 대한 분류 데이터 생성

```python
# 타겟 데이터
fish_target = [1] * 35 + [0] * 14
print(fish_target)
print(len(fish_target))
```

- 도미 1, 빙어 0 으로 분류

#### KNN: K-최근접 이웃 알고리즘 사용
- 사이킷런 모듈 추가

```python
from sklearn.neighbors import KNeighborsClassifier
```

- knn은 모델 객체

```python
knn = KNeighborsClassifier() # 객체 생성
```

- fit() 함수로 `훈련(training)` 시작

```python
# X : 특성데이터, Y : 결과 데이터
knn.fit(fish_data, fish_target)
```

- 훈련 점수 도출

```python
# 훈련시킨 데이터를 그대로 집어넣어서 결과가 제대로 나오는지 확인
# fish_data: 입력 데이터(features), fish_target: 타겟 레이블(labels)
knn.score(fish_data, fish_target)
```

#### K-최근접 이웃 알고리즘
- 주어진 데이터 내에서 새로운 데이터의 근처에 있는 데이터가 무엇인지 확인해서 값을 도출하는 방식

- predict()로 예측 수행

```python
# 길이 30cm, 무게 600g 물고기가 어떤 종류인지 분류 예측
print(knn.predict([[30, 600]]))
print(knn.predict([[15, 200]]))
```

- 위의 데이터는 도미라고 예측
